```python
import tensorflow as tf
tf.__version__
```

```python
from tensorflow.keras.layers import Dense, Flatten, Lambda, Input 
from tensorflow.keras.models import Model, Sequential 
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications.vgg16 import VGG16 
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img
```

```python
from tensorflow.keras.datasets import cifar100
```

```python
(X_train, y_train), (X_test, y_test) = cifar100.load_data()
```

```python
print('Shape of Dataset: ')
print(X_train.shape,'\t',X_test.shape,'\t',y_train.shape,'\t',y_test.shape)
```

```python
image_size = [224, 224]
vgg16 = VGG16(input_shape = image_size + [3], weights = 'imagenet', include_top = False)
resnet50 = ResNet50(input_shape = image_size + [3], weights = 'imagenet', include_top = False)
```

```python
for layer in vgg16.layers:
  layer.trainable = False

for layer in resnet50.layers:
  layer.trainable = False
```

```python
v = Flatten()(vgg16.output)
r = Flatten()(resnet50.output)
```

```python

vgg_prediction = Dense(1000, activation='softmax')(v)
# create a model object
vgg_model = Model(inputs=vgg16.input, outputs=vgg_prediction)

resnet_prediction = Dense(1000, activation='softmax')(r)
# create a model object
resnet_model = Model(inputs=resnet50.input, outputs=resnet_prediction)
```

```python
vgg_model.summary()
```

```python
resnet_model.summary()
```

```python
# Model Compilation
vgg_model.compile(optimizer='adam',loss='categorical_crossentropy',metrics = ['accuracy'])
resnet_model.compile(optimizer='adam',loss='categorical_crossentropy',metrics = ['accuracy'])
```

```python
vgg = vgg_model.fit(X_train, y_train,epochs=20)
```
